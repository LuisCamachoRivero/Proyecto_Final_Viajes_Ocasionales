```python
# Your code here

import pandas as pd
```

```python
df_viajes_ocasionales = pd.read_csv('https://www.sernatur.cl/wp-content/uploads/2020/11/Viajes-ocasionales-1.csv',sep=';')
df_viajes_ocasionales
```

```python
meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre','diciembre']

for mes in meses:
    df_viajes_ocasionales[mes] = df_viajes_ocasionales[mes].str.replace(',', '.').astype(float)

df_viajes_ocasionales
```

```python
df_viajes_ocasionales.info()
```

```python
df_region_count = df_viajes_ocasionales['Region Origen '].value_counts().reset_index()

df_region_count.columns = ['Region Origen', 'count']
df_region_count
```
